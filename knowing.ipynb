```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("./ignore/data_055.csv", encoding="cp949")

df
```

```python
df.keys()
```

```python
# 성별
male = 0
female = 0

for i in df.index.tolist():
    if df.loc[i, "P1"] == "F":
        female += 1
    else:
        male += 1
    

x = np.arange(2)
x_axis = ['male', 'female']
y_axis = [male, female]

plt.bar(x, y_axis)
plt.xticks(x, x_axis)

plt.show()
```

```python
# 고객 이용 연령층
early_20s = 0
lately_20s = 0
early_30s = 0
lately_30s = 0
early_40s = 0
lately_40s = 0
early_50s = 0
lately_50s = 0
early_60s = 0
lately_60s = 0
else_thing = 0

for i in df.index.tolist():
    if (df.loc[i, "P2"] == "20대_초"):
        early_20s += 1
    elif (df.loc[i, "P2"] == "20대_후"):
        lately_20s += 1
    elif (df.loc[i, "P2"] == "30대_초"):
        early_30s += 1
    elif (df.loc[i, "P2"] == "30대_후"):
        lately_30s += 1
    elif (df.loc[i, "P2"] == "40대_초"):
        early_40s += 1
    elif (df.loc[i, "P2"] == "40대_후"):
        lately_40s += 1
    elif (df.loc[i, "P2"] == "50대_초"):
        early_50s += 1
    elif (df.loc[i, "P2"] == "50대_후"):
        early_50s += 1
    elif (df.loc[i, "P2"] == "60대_초"):
        early_60s += 1
    elif (df.loc[i, "P2"] == "60대_후"):
        lately_60s += 1
    else:
        else_thing += 1

x = np.arange(10)
x_axis = ['20s_1', '20s_2', '30s_1', '30s_2', '40s_1', '40s_2', '50s_1', '50s_2', '60s_1', '60s_2']
y_axis = [early_20s, lately_20s, early_30s, lately_30s, early_40s, lately_40s, early_50s, lately_50s, early_60s, lately_60s]

plt.bar(x, y_axis)
plt.xticks(x, x_axis)

plt.show()
```
